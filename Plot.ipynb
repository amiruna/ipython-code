### Large eigenvalue

### a.

```python
import numpy.random as rd
def unif_mat(N):
    M = [np.zeros(N) for a in range(N)]
    M = np.array(M)
    for i in range(N):
        for j in range(0,i):
            M[i][j] = M[j][i]
        for j in range(i, N):
            M[i][j] = rd.uniform()
    return M
```

```python
M = unif_mat(30)
```

```python
eigs = np.linalg.eigh(np.array(M))
```

```python
print max(eigs[0])
```

```python
x = np.arange(10, 520, 20)
y = [max(np.linalg.eigh(unif_mat(n))[0]) for n in x]
```

```python
plt.scatter(x,y, color = 'red', s = 30, zorder = 1)
plt.plot(x,y, color = 'k', zorder = 0, lw = 1)
plt.ylabel("Largest eigenvalue")
plt.xlabel("N (size of the matrix)")
plt.xlim(0,520)
plt.ylim(0,260)
plt.show()
```

### Level Repulsion

### b.

```python
def norm_mat(N):
    M = [np.zeros(N) for a in range(N)]
    M = np.array(M)
    for i in range(N):
        for j in range(0,i):
            M[i][j] = M[j][i]
        for j in range(i, N):
            M[i][j] = rd.normal()
    return M
```

```python
M = norm_mat(3000)
```

```python
eigs = np.linalg.eigh(np.array(M))
```

```python
rcParams['figure.figsize'] = (10, 6)
plt.hist(eigs[0], bins = 50, normed=True)
plt.xlim(-110,110)
plt.xlabel("Eigenvalue")
plt.ylabel("Frequency")
plt.show()
```

```python
l = eigs[0]
```

```python
unfolded_spacings = l[1:] - l[:-1]
```

```python
from scipy.integrate import quad
def integrand(x, N):
    return 1/2/np.pi/N*np.sqrt(4*N-x**2)
def circle_cdf(N, x):
    I = quad(integrand, -2*np.sqrt(N), x, args=(N))
    return I[0]
```

```python
lambda_cdf = np.array([circle_cdf(3000, x) for x in l])
spacings = lambda_cdf[1:] - lambda_cdf[:-1]
```

```python
#Wigner's surmise
def wigner(N, s):
    return N**2*np.pi/2*s*np.exp(-np.pi/4*s**2*N**2)

x = np.arange(0, 0.001, 0.00002)
y = [wigner(3000, a) for a in x]
```

```python
rcParams['figure.figsize'] = (16, 6)
plt.figure(1)
plt.subplot(1,2,1)
plt.hist(eigs[0], bins = 50, normed=True)
plt.xlim(-110,110)
plt.xlabel("$\lambda$")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(lambda_cdf, bins = 30, normed = True)
plt.xlabel("C($\lambda$)")
plt.show()
```

```python
plt.hist(spacings, bins = 30, normed = True, label = "Level spacings")
plt.plot(x,y, label = "Wigner's surmise")
plt.ylabel("Frequency")
plt.xlabel("Spacing")
plt.legend()
plt.show()
```

### c-d.

```python
def norm_sparse_mat(N, f):
    M = [np.zeros(N) for a in range(N)]
    M = np.array(M)
    for i in range(N):
        for j in range(0,i):
            M[i][j] = M[j][i]
        for j in range(i, N):
            M[i][j] = (rd.binomial(1, f)) * rd.normal()
    return M
```

```python
M = norm_sparse_mat(3000, 0.001)
```

```python
eigs = np.linalg.eigh(np.array(M))
```

```python
l = eigs[0]
```

```python
lambda_cdf = np.array([circle_cdf(3000, x) for x in l])
spacings = lambda_cdf[1:] - lambda_cdf[:-1]
```

```python
x = np.arange(0, 0.001, 0.00002)
y = [wigner(3000, a) for a in x]
```

```python
rcParams['figure.figsize'] = (16, 6)
plt.figure(1)
plt.subplot(1,2,1)
plt.hist(eigs[0], bins = 50, normed=True)
plt.xlabel("$\lambda$")
plt.ylabel("Frequency")
plt.title("f = 0.001")

plt.subplot(1,2,2)
plt.hist(lambda_cdf, bins = 30, normed = True)
plt.xlabel("C($\lambda$)")
plt.show()
```

```python
rcParams['figure.figsize'] = (10, 6)
plt.hist(spacings, bins = 40, normed = True, label = "Level spacings")
plt.plot(x,y, label = "Wigner's surmise")
plt.ylabel("Frequency")
plt.xlabel("Spacing")
plt.legend()
plt.title("f = 0.001")
plt.show()
```

```python
def complex_mat(N):
    M = [np.zeros(N)*1j for a in range(N)]
    M = np.array(M)
    for i in range(N):
        for j in range(0,i):
            M[i][j] = np.conj(M[j][i])
        for j in range(i, N):
            M[i][j] = rd.normal()+rd.normal()*1j
    return M
```

```python
M = complex_mat(3000)
```

```python
eigs = np.linalg.eigh(np.array(M))
l = eigs[0]
```

```python
lambda_cdf = np.array([circle_cdf(6000, x) for x in l])
spacings = lambda_cdf[1:] - lambda_cdf[:-1]
```

```python
x = np.arange(0, 0.001, 0.00002)
y = [wigner(3000, a) for a in x]
```

```python
rcParams['figure.figsize'] = (16, 6)
plt.figure(1)
plt.subplot(1,2,1)
plt.hist(eigs[0], bins = 50, normed=True)
plt.xlabel("$\lambda$")
plt.ylabel("Frequency")
plt.title("Non-sparse")

plt.subplot(1,2,2)
plt.hist(spacings, bins = 40, normed = True, label = "Level spacings")
plt.plot(x,y, label = "Wigner's surmise")
plt.ylabel("Frequency")
plt.xlabel("Spacing")
plt.legend()
plt.show()
```

```python
def sparse_complex_mat(N,f):
    M = [np.zeros(N)*1j for a in range(N)]
    M = np.array(M)
    for i in range(N):
        for j in range(0,i):
            M[i][j] = np.conj(M[j][i])
        for j in range(i, N):
            M[i][j] = (rd.binomial(1, f))*(rd.normal()+rd.normal()*1j)
    return M
```

```python
M = sparse_complex_mat(3000,0.001)
eigs = np.linalg.eigh(np.array(M))
l = eigs[0]
lambda_cdf = np.array([circle_cdf(6000, x) for x in l])
spacings = lambda_cdf[1:] - lambda_cdf[:-1]
x = np.arange(0, 0.001, 0.00002)
y = [wigner(3000, a) for a in x]
```

```python
rcParams['figure.figsize'] = (16, 6)
plt.figure(1)
plt.subplot(1,2,1)
plt.hist(eigs[0], bins = 50, normed=True)
plt.xlabel("$\lambda$")
plt.ylabel("Frequency")
plt.title("f = 0.001")

plt.subplot(1,2,2)
plt.hist(spacings, bins = 40, normed = True, label = "Level spacings")
plt.plot(x,y, label = "Wigner's surmise")
plt.xlim(0, 0.0007)
plt.ylabel("Frequency")
plt.xlabel("Spacing")
plt.legend()
plt.show()
```
