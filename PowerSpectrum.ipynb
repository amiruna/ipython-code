### b. Correlation function with cyclic boundary conditions

```python
from scipy import signal 

data_length = len(wave[1])

a = wave[1][:data_length]
a = a/max(a)
b = np.zeros(data_length * 2)
b[data_length/2:data_length/2+data_length] = a 

# array flipped convolution == which is a correlation.
c = signal.fftconvolve(b, a[::-1], mode='valid') 
c = c[c.size/2:]
```

```python
plt.plot(range(len(c)), c)
plt.ylim(-5000,5000)
plt.xlim(0,20000)
plt.show()
```

```python
plt.plot(range(len(c)), c)
plt.ylim(-10000,10000)
plt.xlim(0,800)
plt.show()
```

### c. The power‐spectrum, by averaging over several windows. 

```python
from scipy.signal import hanning
```

```python
length = int(len(wave[1])/200)*200
a = wave[1][:length]
noWindows = 2*100-1
windowSize = length/100
halfWindowSize = windowSize/2

b = np.empty([noWindows, windowSize], dtype=complex)
for i in range(noWindows):
    b[i,:] = a[i*halfWindowSize:i*halfWindowSize+windowSize]
    
res = np.empty([noWindows, windowSize], dtype=complex)
```

```python
w = hanning(windowSize)
for i in range(noWindows):
    res[i,:] = scipy.fftpack.fft(b[i,:] * w)
    res[i,:] = [np.abs(res[i][j])**2/windowSize for j in range(int(windowSize))] 
```

```python
res = res.transpose()
means = [np.mean(res[i,:]) for i in range(len(res))]
```

```python
plt.plot(range(len(means)), means)
plt.xlim(0,1500)
plt.show()
```

```python
plt.plot(range(len(means)), means)
plt.xlim(0,1500)
plt.show()
```

### d.  The power‐spectrum, by smoothing the Fourier‐transform using an appropriate convolution. 

```python
N = np.log2(len(wave[1]))
N = 2**(int(N)+1)
fft_smooth = np.fft.fft(wave[1],N)
N = np.ceil((N+1)/2)
fft_smooth = fft_smooth[:N]
ps_smooth = [np.abs(a)**2 for a in fft_smooth]
ps_smooth = np.array(ps_smooth)/len(wave[1])**2
```

```python
plt.plot(range(len(ps_smooth)), ps_smooth)
plt.show()
```

```python
plt.plot(range(len(ps_smooth)), ps_smooth)
plt.xlim(0,130000)
plt.ylim(0,2100)
plt.show()
```
