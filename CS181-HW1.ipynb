```python
data = read_csv('motorcycle.csv')
axis_names = data.columns
data.columns = ['time','force']
```

```python
data.head()
```

```python
times = data['time'].values
forces = data['force'].values
```

```python
plt.scatter(times, forces, s=50, facecolor = 'none', edgecolors = 'black')
plt.xlim(0,60)
plt.xlabel(axis_names[0])
plt.ylabel(axis_names[1])
plt.show()
```

```python
'''
takes in function that converts inputs into design matrix
'''
def plot_reg(phi_transform):
    #compute design matrix
    Phi = phi_transform(times)
    
    # Find the regression weights using the Moore-Penrose pseudoinverse.
    w = np.linalg.solve(np.dot(Phi.T, Phi) , np.dot(Phi.T, forces))
    
    # Compute the regression line on a grid of inputs.
    grid_times = np.linspace(0, 60, 200)
    grid_Phi     = phi_transform(grid_times)
    grid_Yhat  = np.dot(grid_Phi, w)
    
    # Plot the data and the regression line.
    plt.plot(times, forces, 'o',
             grid_times, grid_Yhat, '-')
    plt.ylim(-150, 100)
    plt.xlabel(axis_names[0])
    plt.ylabel(axis_names[1])
    plt.show()
```

### Linear basis

```python
#for testing purposes
def lin_phi(x):
    return np.vstack((np.ones(x.shape), x)).T

plot_reg(lin_phi)
```

### Other basis

### a) $\phi_j(x)=\exp\{-((x-10j)/5)^2\}, \; j = 0,...,6$

```python
'''
function takes a numeric factor, returns Phi
'''
def exp_basis(x, a):
    X = np.ones(x.shape)
    for j in range(7):
        X = np.vstack((X, np.exp(-((x - 10*j)/a)**2)))
    #return X[1:].T
    return X.T

def phi_wrapper(a):
    return partial(exp_basis, a = a)
```

```python
plot_reg(phi_wrapper(5))
```

### b) $\phi_j(x)=\exp\{-((x-10j)/10)^2\}, \; j = 0,...,6$

```python
plot_reg(phi_wrapper(10))
```

### c) $\phi_j(x)=\exp\{-((x-10j)/25)^2\}, \; j = 0,...,6$

### d) $\phi_j(x) = x^j, \; j = 0,...,10$

```python
def poly_basis(x):
    X = np.ones(x.shape)
    for j in range(11):
        X = np.vstack((X, x**j))
    return X[1:].T
```

```python
plot_reg(poly_basis)
```

### e) $\phi_j(x) = \sin \left(\frac{x}{j}\right), \; j = 1,...,20$

```python
def sine_basis(x):
    X = np.ones(x.shape)
    for j in range(1,21):
        X = np.vstack((X, np.sin(x/j)))
    return X.T
```

```python
plot_reg(sine_basis)
```
