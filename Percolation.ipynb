```python
#Depth first search
def search_dfs(G, i, visited):
    visited[i]=1
    for node in G[i]:
        if visited[node]==0:
            search_dfs(G, node, visited)

def DFS(G, N):
    visited = {}
    for node in G.keys():
        visited[node]=0
    for node in G.keys():
        if node%(N+1)==0 and visited[node]==0:
            search_dfs(G, node, visited)
    for node in G.keys():
        if node%(N+1)==N and visited[node] == 1:
            return 1
    return 0
```

```python
#Breadth first search without recursion
def search_bfs(G, i, visited):
    visited[i] = 1
    neighbors = G[i]
    while neighbors:
        node = neighbors.pop()
        if not visited[node]:
            visited[node] = 1
            neighbors+=G[node]
        
def BFS(G, N):
    visited = {}
    for node in G.keys():
        visited[node]=0
    for node in G.keys():
        if node%(N+1)==0 and visited[node]==0:
            search_bfs(G, node, visited)
    for node in G.keys():
        if node%(N+1)==N and visited[node] == 1:
            return 1
    return 0
```

```python
#Simulate bond percolation on a lattice of size N
def sim_perc(N, p):
    h_bonds = rd.binomial(1, p, (N+1, N))
    v_bonds = rd.binomial(1, p, (N, N+1))
    
    for r in range(N+1):
        for i in range(N):
            if h_bonds[r][i]:
                plt.hlines(r, i, i+1, alpha = 1, lw = 4.5, color = 'k')
                
    for r in range(N):
        for i in range(N+1):
            if v_bonds[r][i]:
                plt.vlines(i, r, r+1, alpha = 1, lw = 4.5, color = 'k')
                
    #plt.title("N = "+str(N)+", p = "+str(p)+"\n")
```

### b. Calculations: percolation probability

```python
def is_perc(N, p):
    h_bonds = rd.binomial(1, p, (N+1, N))
    v_bonds = rd.binomial(1, p, (N, N+1))
    
    G = {}
    for idx in range(0, (N+1)**2):
        G[idx] = []
        i = idx/(N+1)
        j = idx%(N+1)
        
        if j < N and h_bonds[i][j]:
            G[idx].append(idx+1)
        if j >= 1 and h_bonds[i][j-1]:
             G[idx].append(idx-1)
        if i < N and v_bonds[i][j]:
            G[idx].append(idx+(N+1))
        if i >= 1 and v_bonds[i-1][j]:
             G[idx].append(idx-(N+1))
     
    return BFS(G, N)
```

```python
def plot_pc(a):
    rcParams['figure.figsize'] = (8, 6)
    x = np.arange(0,1.01,0.01)
    j = 0
    for N in a:
        y = []
        for p in x:
            sims = np.array([is_perc(N, p) for i in range(50)])
            y.append(np.mean(sims))
        plt.plot(x,y, color = dark2_colors[j], zorder = 0, label = "N = "+str(N))
        plt.scatter(x,y, color = 'k', zorder = 1, s = 15)
        plt.ylim(-0.1, 1.1)
        plt.xlim(0.4, 0.6)
        plt.xticks(np.arange(0.4,0.7,0.1))
        j+=1
    plt.hlines(1/3, 0.4, 0.6, alpha = 1, lw = 1, linestyle = '--')
    plt.hlines(2/3, 0.4, 0.6, alpha = 1, lw = 1, linestyle = '--')
    plt.xlabel("p")
    plt.ylabel("$p_c$")
    plt.legend(loc = 'upper left')
    
```

```python
def trans_width(a):
    rcParams['figure.figsize'] = (8, 6)
    x = np.arange(0,1.01,0.025)
    
    #window width
    w = []
    
    for N in a:
        print N
        #temporary widths for each simulation
        widths = []
        for j in range(10):
            y = []
            for p in x:
                sims = np.array([is_perc(N, p) for i in range(50)])
                y.append(np.mean(sims))
        
            #fit a line and find the width 
            l = 0
            r = len(x)-1
            for i in range(len(x)):
                if y[i] > 1/3:
                    l = i
                    break
            for i in np.arange(len(x)-1,-1,-1):
                if y[i] < 2/3:
                    r = i
                    break
            slope = np.polyfit(x[(l-1):(r+2)], y[(l-1):(r+2)], 1)[0]
            widths.append(1/3/slope)
        w.append(np.mean(widths))
    
    #plot transition width as a function of 
    a = np.array(a)
    plt.plot(np.log(a), np.log(w))
    plt.scatter(np.log(a), np.log(w))
    plt.xlabel("$\log N$")
    plt.ylabel("$\log w$")
    plt.plot()
    print np.polyfit(np.log(a), np.log(w), 1)[0]    
```
