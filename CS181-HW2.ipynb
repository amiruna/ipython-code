### 1. Warm-up: Gaussian generative classifier

```python
class_prob = np.array([n_apples/N, n_orange/N, n_lemon/N])
```

```python
apples_idx = 0
orange_idx = n_apples
lemon_idx = n_apples + n_orange
all_N = [n_apples, n_orange, n_lemon]
```

```python
#compute class means
class_means = []
for i in range(2):
    var = data[data.columns[i+1]].values
    mu = [np.mean(var[:orange_idx]),np.mean(var[orange_idx:lemon_idx]), np.mean(var[lemon_idx:])]
    class_means.append(mu)
class_means = np.array(class_means)
```

```python
class_means = class_means.T
```

```python
class_means
```

```python
#compute the covariance matrix
S = []
for i in range(3):
    S_i = np.zeros((2,2))
    for n in range(N):
        vec = (X[n]-class_means[i])[np.newaxis]
        S_i += T[n][i]*np.dot(vec.T, vec)
    S_i /= N
    S.append(S_i)
```

```python
Sigma = sum(S)
Sigma
Sigma_inv = np.linalg.inv(Sigma)
```

```python
w = [np.dot(Sigma_inv, class_means[i]) for i in range(len(class_means))]
```

```python
aux = [-0.5*np.dot(Sigma_inv, ((class_means[i])[np.newaxis]).T) for i in range(len(class_means))]
```

```python
w0 = [np.dot(class_means[i], aux[i])[0]+np.log(class_prob[i]) for i in range(len(class_means))]
```

```python
w0
```

```python
w
```

```python
def posterior_prob(x):
    prob = []
    for k in range(3):
        prob.append(np.exp(sum(w[k]*x)+w0[k]))
    return np.array(prob)/sum(prob)
```

```python
posterior_prob(X[1])
```

```python
my_range = np.arange(4,12,0.02)
points = np.array([np.array([i,j]) for i in my_range for j in my_range])
```

```python
x = points.T[0]
y = points.T[1]
colors = ['red', 'orange', 'yellow']
klass = [colors[np.argmax(posterior_prob(pt))] for pt in points]
```

```python
rcParams['axes.grid'] = False
for color in colors:
    x1 = []
    y1 = []
    for i in range(len(klass)):
        if klass[i]==color:
            x1.append(x[i])
            y1.append(y[i])
    plt.plot(x1,y1,color = color, zorder = 0)
    
plt.scatter(data['width'][:n_apples], data['height'][:n_apples], s = 100, color = 'red', 
            edgecolor = 'k', label="Apples", zorder = 1)
plt.scatter(data['width'][n_apples:(n_apples+n_orange)], data['height'][n_apples:(n_apples+n_orange)], s = 100, 
            color = 'orange', edgecolor = 'k', label = "Oranges", zorder = 1)
plt.scatter(data['width'][(n_orange+n_apples):], data['height'][(n_orange+n_apples):], s = 100, 
            color = 'yellow', edgecolor = 'k', label = "Lemons", zorder = 1)
plt.xlabel("Width (cm)")
plt.ylabel("Height (cm)")
plt.title("Decision Boundaries Using Gaussian Class-conditional")
plt.legend(loc = 'lower right')
plt.xlim(4,12)
plt.ylim(4,12)
plt.show()
```
